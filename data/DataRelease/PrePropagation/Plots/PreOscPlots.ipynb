```python
%matplotlib inline
import pylab
import numpy
import os
```

```python
Files=os.listdir("../")
Nus=[r"$\nu_e$",r"$\nu_\mu$",r"$\nu_\tau$"]
FNames=['e','mu','tau']
for inu in range(0,3):
    pylab.figure(figsize=(10,5))
    for f in Files:
        if("SIBYLL" in f):
            dat=numpy.loadtxt("../"+f+"/flux_b0.0.txt")
            pylab.loglog(dat[:,0],dat[:,0]*dat[:,0]*dat[:,0]*dat[:,inu+1],label=f)
            pylab.ylim(1e-8, 1e0)
            pylab.xlim(10,1e19)
            pylab.title(Nus[inu]+" at b=0",fontsize=22)
    pylab.xlabel(r"$E_\nu$ / GeV",fontsize=16)
    pylab.ylabel(r"$\Phi_{\nu}$ ($E_\nu$/GeV)$^3$ (cm$^{2}$ s sr GeV)$^{-1}$",fontsize=16)
    pylab.legend(loc='right',fontsize=11.5)
    pylab.grid(which='both')
    pylab.tight_layout()
    pylab.savefig("SummaryPlot_nu_"+FNames[inu]+".pdf")
    pylab.savefig("SummaryPlot_nu_"+FNames[inu]+".png")
    pylab.show()
```

```python
Files=os.listdir("../")
Nus=[r"$\bar{\nu}_e$",r"$\bar{\nu}_\mu$",r"$\bar{\nu}_\tau$"]
FNames=['e','mu','tau']
for inu in range(0,3):
    pylab.figure(figsize=(10,5))
    for f in Files:
        if("SIBYLL" in f):
            dat=numpy.loadtxt("../"+f+"/flux_0.0_b0.0000_E.txt")
            pylab.loglog(dat[:,0],dat[:,0]*dat[:,0]*dat[:,0]*dat[:,inu+4],label=f)
            pylab.ylim(1e-8, 1e0)
            pylab.xlim(10,1e19)
            pylab.title(Nus[inu]+" at b=0",fontsize=22)
    pylab.xlabel(r"$E_\nu$ / GeV",fontsize=16)
    pylab.ylabel(r"$\Phi_{\nu}$ ($E_\bar{\nu}$/GeV)$^3$ (cm$^{2}$ s sr GeV)$^{-1}$",fontsize=16)
    pylab.legend(loc='right',fontsize=11.5)
    pylab.grid(which='both')
    pylab.tight_layout()
    pylab.savefig("SummaryPlot_antinu_"+FNames[inu]+".pdf")
    pylab.savefig("SummaryPlot_antinu_"+FNames[inu]+".png")
    pylab.show()
```
